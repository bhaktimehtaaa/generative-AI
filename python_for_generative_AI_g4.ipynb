```python
import numpy as np
from numpy.linalg import norm

A = np.array([2,1,2,3,2,9])
B = np.array([3,4,2,4,5,5])

#compute cosine similarity
cosine = np.dot(A,B) / (norm(A) * norm(B))
print("Cosine Similarity:", cosine)
```

```python
from transformers import pipeline

summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

text = """the global shift toward renewable energy is accelerating,
driven by concerns over climate change and energy security. solar
and wind pipe installastions have reached record levels worlwide."""

summary = summarizer(text, max_length=30, min_length=10, do_sample=False)
print(summary[0])
```

```python
from transformers import AutoTokenizer

# Load a tokenizer (e.g., BERT)
tokenizer = AutoTokenizer.from_pretrained("bert-bert-uncased")

text = "Machine learning is transforming materials science!"

#Tokenize
tokens = tokenizer.tokenize(text)
tokens_ids = tokenizer.encode(text)

print("Tokens:", tokens)
print("Token_ids")
```
